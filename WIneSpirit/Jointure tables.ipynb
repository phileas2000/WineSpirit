```python
#fusion des tables
cursor.execute("SELECT * FROM sqlite_master")
cursor.execute("DROP TABLE IF EXISTS erpWeb")
cursor.execute("""CREATE TABLE erpWeb AS SELECT * FROM web INNER JOIN liaison ON id_web = sku INNER JOIN erp ON erp.product_id = liaison.product_id""")
```

```python
print(pd.read_sql_query("SELECT id_web,product_id, COUNT(*) c FROM liaison GROUP BY id_web HAVING c > 1",conn))
print(pd.read_sql_query("SELECT product_id, COUNT(*) c FROM liaison GROUP BY product_id HAVING c > 1",conn))
print(pd.read_sql_query("SELECT sku, COUNT(*) c FROM web GROUP BY sku HAVING c > 1",conn))
#pas de doublons
```

```python
print(pd.read_sql_query("SELECT product_id, COUNT(*) c FROM erpWeb GROUP BY product_id HAVING c > 1",conn))
#détection des product_id doublon
print(pd.read_sql_query("SELECT * FROM erpWeb WHERE product_id= 3847",conn))
```

```python
print(pd.read_sql_query("SELECT product_id,id_web FROM erpWeb WHERE id_web IS NULL OR product_id IS NULL ",conn))
#verification absence de correspondances
```
